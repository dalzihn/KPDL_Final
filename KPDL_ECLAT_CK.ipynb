```python
file_path = '/content/Books.csv'
```

```python
import pandas as pd
df = pd.read_csv(file_path)
print(df.head())
```

```python
# Chuyển đổi dữ liệu sang horizational format
result = df.apply(lambda row: [col for col, val in row.items() if val == 1] + [None] * (len(df.columns) - sum(row)), axis=1)

# Chuyển đổi kết quả về dataframe mới
result_df = pd.DataFrame(result.tolist(), columns=df.columns)

# Hiển thị kết quả
print(result_df)
```

```python
# prompt: code python để chuyển header thành số 0,1,2,3

# Chuyển đổi dữ liệu sang horizational format
result = df.apply(lambda row: [col for col, val in row.items() if val == 1] + [None] * (len(df.columns) - sum(row)), axis=1)

# Chuyển đổi kết quả về dataframe mới
result_df = pd.DataFrame(result.tolist(), columns=df.columns)

# Thay đổi header thành số 0, 1, 2, 3,...
result_df.columns = range(len(result_df.columns))

# Hiển thị kết quả
result_df
```

```python
# prompt: replace None thành NaN
# Replace None with NaN
result_df = result_df.replace({None: np.nan})

# Hiển thị kết quả
result_df
```

```python
from pyECLAT import ECLAT
eclat_instance = ECLAT(data=result_df,verbose=True)
```

```python
# the item shoud appear at least at 5% of transactions
min_support = 22/100
# start from transactions containing at least 2 items
min_combination = 2
# up to maximum items per transaction
max_combination = result_df.apply(lambda row: sum(row.notna()), axis=1).max()
rule_indices, rule_supports = eclat_instance.fit(min_support=min_support,
                                                 min_combination=min_combination,
                                                 max_combination=max_combination,
                                                 separator=' & ',
                                                 verbose=True)
```

```python
import pandas as pd
result = pd.DataFrame(rule_supports.items(),columns=['Item', 'Support'])
result.sort_values(by=['Support'], ascending=False)
```
